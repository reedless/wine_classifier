```python
import pandas as pd
df = pd.read_csv("wine_quality.csv")
```

```python
df.head()
```

```python
print("Shap" + df.shape)


classes = df.country.unique()

df_accum = (df[df.country==classes[0]].describe().drop(columns=['points', 'price']))

for country in classes:
    df_accum = (df_accum.join(df[df.country==country].describe().round(1)
                               .rename(columns={'points': 'points '+country, 'price':'price '+country})))
df_accum
```
